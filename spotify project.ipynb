```python
corr_df = dataset.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,vmin=-1,vmax=1,center=0,cmap="crest",linewidths=1,linecolor="black")
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

# this is analysis of track duration using frequency distribution chart

```python
X = dataset.rename(columns={'duration_ms' : 'duration_s'})
X=X['duration_s'].loc[X['duration_s'] < 800000] / 10**3

# plotting
plt.figure(figsize=(20, 5))
sns.distplot(X)
plt.show()
```
